```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Baca gambar
image = cv2.imread('saya.jpg')
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Konversi ke grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Pencerahan (brightness adjustment)
bright_gray = cv2.convertScaleAbs(gray, alpha=1, beta=150)  # alpha=1 (konstan), beta=+50 (pencerahan)

# Penambahan kontras (contrast adjustment)
contrast_gray = cv2.convertScaleAbs(gray, alpha=1.5, beta=0)  # alpha>1 meningkatkan kontras, beta=0

# Kombinasi pencerahan dan kontras
bright_contrast_gray = cv2.convertScaleAbs(gray, alpha=1.5, beta=150)

# Tampilkan hasil
titles = ['Original Image', 'Grayscale', 'Brightened Gray', 'Contrast Gray', 'Bright + Contrast Gray']
images = [image_rgb, gray, bright_gray, contrast_gray, bright_contrast_gray]

plt.figure(figsize=(15, 10))
for i in range(5):
    plt.subplot(2, 3, i + 1)
    cmap = 'gray' if len(images[i].shape) == 2 else None
    plt.imshow(images[i], cmap=cmap)
    plt.title(titles[i])
    plt.axis('off')
plt.tight_layout()
plt.show()
```
